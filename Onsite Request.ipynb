```python
class Onsite:
    def __init__ (self, account, name, departing, returning, departure, destination, transportation):
        self.account = account
        self.name = name
        self.departing = departing
        self.returning = returning
        self.departure = departure
        self.destination = destination
        self.transportation = transportation
```

```python
from tkinter import *
from tkinter.ttk import *
from tkinter import messagebox
from tkcalendar import Calendar, DateEntry
window = Tk()
window.title("Onsite Request")
list_onsite = [] #Lưu trữ danh sách các yêu câu công tác

# tạo các label 
account_lbl = Label(window, text="Account")
name_lbl = Label(window, text="Name")
departing_lbl = Label(window, text="Departing Date")
returning_lbl = Label(window, text="Returning Date")
departure_lbl = Label(window, text="Departure")
destination_lbl = Label(window, text="Destination")
transportation_lbl = Label(window, text="Transportation")
searching_lbl = Label(window, text="You can search your request here!")

# tạo các textbox nhập dữ liệu
account = Entry(window)
name = Entry(window)
departing= Entry(window)
returning = Entry(window)
departure = Entry(window)
destination = Entry(window)
search = Entry(window)

# tạo ô nhập ngày tháng năm
departing= DateEntry(window, width=18, background='darkblue',foreground='white', borderwidth=2, year=2022)
returning= DateEntry(window, width=18, background='darkblue',foreground='white', borderwidth=2, year=2022)

# tạo combobox nhập loại hình chuyến bay
transportation = Combobox(window, width = 18)
transportation['value'] = ('Flight','Bus', 'Train')
transportation.current(0)

# hàm thêm request
def add_request():
    if departing.get() <= returning.get():
        list_onsite.append(Onsite(account.get(), name.get(), departing.get(), returning.get(), departure.get(), destination.get(), transportation.get()))
        print([item.account for item in list_onsite], "has a business trip from", [item.departure for item in list_onsite], [item.departing for item in list_onsite], "to", [item.destination for item in list_onsite], [item.returning for item in list_onsite])
        messagebox.showinfo("Onsite Request", account.get() + " " + name.get() + " added sucessfully")
    else:
        messagebox.showinfo("Onsite Request", "Returning Date is not valid")
    
# hàm tìm kiếm request
def search_request():
    search_account = search.get()
    exist = False #giả định item ko tồn tại
    for item in list_onsite:
        #đã có kết quả tìm kiếm phù hợp
        if item.account == search_account: 
            exist = True
            # set các giá trị cho các label
            account.delete(0,"end")
            name.delete(0,"end")
            departing.delete(0,"end")
            returning.delete(0,"end")
            departure.delete(0,"end")
            destination.delete(0,"end")
            account.insert(0,item.account)
            name.insert(0,item.name)
            departing.insert(0,item.departing)
            returning.insert(0,item.returning)
            departure.insert(0,item.departure)
            destination.insert(0,item.destination)
            # xuất thông báo
            break
    if exist:
        messagebox.showinfo("Onsite Request", f"{item.name} onsite from {item.departure} {item.departing} to {item.destination} {item.returning}")
        # kết quả tìm kiếm không phù hợp
    else: 
        messagebox.showinfo("Onsite", f"{search_account} does not exist in list request")

#Hàm xóa request
def delete_request():
    delete_account = account.get()
    exist = False
    # duyệt từng chuyến bay trong list_onsite
    for i in range(len(list_onsite)):
        if list_onsite[i].account == delete_account: #Nếu phần tử đang duyệt giống với phần tử đang hiển thị
            exist = True
            list_onsite.remove(list_onsite[i])
            messagebox.showinfo("Onsite request", f"{delete_account} delete request sucessfully")
            print([item.account for item in list_onsite], "has a business trip from", [item.departure for item in list_onsite],
             [item.departing for item in list_onsite], "to", [item.destination for item in list_onsite], [item.returning for item in list_onsite])

        else: # Ngược lại, xuất thông báo delete không thành công
            messagebox.showinfo("Onsite request", f"{delete_account} does not exist in list request")

# Hàm cập nhật request
def update_request():
    update_account = search.get()
    update_departing = departing.get()
    update_returning = returning.get()
    exist = False # giả định item không tồn tại trong list
    for i in range(len(list_onsite)):
        if list_onsite[i].account == update_account:
            exist = True
            list_onsite[i].departing=update_departing
            list_onsite[i].returning=update_returning
    print([item.account for item in list_onsite], "updated onsite request from", [item.departure for item in list_onsite], [item.departing for item in list_onsite], "to", [item.destination for item in list_onsite], [item.returning for item in list_onsite])
    messagebox.showinfo("Onsite request", name.get() + " updated sucessfully")       
    
#Tạo button
add_btn = Button(window, text="Add", command=add_request)
delete_btn = Button(window, text="Delete", command=delete_request)
update_btn = Button(window, text="Update", command=update_request)
search_btn = Button(window, text="Search", command=search_request)

# căn chỉnh vị trí các label
account_lbl.grid(row = 1, column = 0)
name_lbl.grid(row = 2, column = 0)
departure_lbl.grid(row = 3, column = 0)
destination_lbl.grid(row = 4, column = 0)
departing_lbl.grid(row = 5, column = 0)
returning_lbl.grid(row = 6, column = 0)
transportation_lbl.grid(row = 7, column = 0)
account.grid(row = 1, column = 1, columnspan = 3)
name.grid(row = 2, column = 1, columnspan = 3)
departure.grid(row = 3, column = 1, columnspan = 3)
destination.grid(row = 4, column = 1, columnspan = 3)
departing.grid(row = 5, column = 1, columnspan = 3)
returning.grid(row = 6, column = 1, columnspan = 3)
transportation.grid(row = 7, column = 1, columnspan = 3)
add_btn.grid(row = 8, column = 0)
delete_btn.grid(row = 8, column = 1)
update_btn.grid(row = 8, column = 2)
searching_lbl.grid (row = 9 , column = 0, columnspan = 4)
search_btn.grid(row = 10, column = 1)
search.grid(row = 10, column = 0)


window.mainloop()
```
